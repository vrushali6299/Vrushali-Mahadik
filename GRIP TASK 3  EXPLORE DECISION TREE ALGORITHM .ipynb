# Importing Essential Libraries

```python
import sklearn.datasets as datasets
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Importing Dataset 

```python
iris=datasets.load_iris()
# Forming the iris dataframe
df=pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
x=df
y=iris.target
```

# Histogram

```python
df.hist(figsize = (10,10)) 
```

# Define the Decision Tree Algorithm

```python
from sklearn .model_selection import train_test_split 
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.20,random_state = 2020)

# Apply Decision tree algorithms 
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.metrics import accuracy_score,confusion_matrix

clf = DecisionTreeClassifier()
clf.fit(x_train,y_train) #fitting model
```

# Checking Accuracy of training and testing data

```python
predict = clf.predict(x_train)
print("Accuracy of training data : ",accuracy_score(predict,y_train)*100,"%")
print("Confusin matrix of training data :'\n' ",confusion_matrix(predict,y_train))
```

```python
# Accuracyon testing data
predict = clf.predict(x_test)
print("Accuracy of testing data : ",accuracy_score(predict,y_test)*100,"%")
print("Confusin matrix of testing data :\n ",confusion_matrix(predict,y_test))
```

# Let us visualize the Decision Tree to understand it better

```python
plt.figure(figsize  = (14,14))
tree.plot_tree(clf,filled = True,rounded = True,proportion = True,node_ids = True , feature_names = iris.feature_names)
plt.show()
```
